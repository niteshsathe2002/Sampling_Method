#### Histogram

```python
plt.hist(data,bins=10,color="maroon",edgecolor="k",alpha=1)
plt.title("Histogram")
plt.show()
```

#### Boxplot

```python
sns.boxplot(data,width=0.5,color="red")
```

```python
table=pd.crosstab(data,'count')
table.head()
```

#### Normal Probability Plot

```python
mean=np.mean(data)
sd=np.std(data)
plt.plot(data,norm.pdf(data,mean,sd))
plt.show()
```

```python
range=np.max(data)-np.min(data)
print("range=",range)
```

```python
inter_qurtile_range=(np.percentile(data,75)-np.percentile(data,25))/2
print("inter_qurtile_range=",inter_qurtile_range)
```
