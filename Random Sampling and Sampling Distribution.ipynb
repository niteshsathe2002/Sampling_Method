#### Random sample from Poisson Disribution (Population)

```python
N=10000
lam=3
population=np.random.poisson(lam,N)
population_mean=np.mean(population)
```

   #### Sample From Population

```python
n=100
k=5
sample_means=[]
for i in range(k):
    sample=np.random.choice(population,n,replace=False)
    sample_mean=np.mean(sample)
    sample_means.append(sample_mean)
    i+=1
    
```

#### Samples mean and Population mean

```python
print("sample_means=",sample_means)
print("population_mean=",np.mean(population))
```

#### Sampling Errors

```python
sampling_errors=population_mean-sample_means
print("sampling_errors=",sampling_errors)
```

**Q3 Draw 10,000 samples of size 25 from Normal (100,${3^2}$) and study the sampling distribution of following statistics for n=5,10**
 $i) Y_{bar}$
  $ii)  S_y=\frac{1}{n-1}\sum_{i=1}^n(Y_r-{Ybar})^2$
  $iii) Y_{(1)}=min(Y_1,Y_2,....Y_n)$
 $ iv) Y_{(n)}=max(Y_1,Y_2,....Y_n)$
  $v) R=Y_{(n)}-Y_{(1)}$
        

```python
n=25
n1=5
n2=10
k=10000
mu=1000
sd=3
samples_mean=[]
samples_var=[]
samples_min=[]
samples_max=[]
samples_range=[]
for i in range(k):
    population=np.random.normal(mu,sd,n)
    sample=np.random.choice(population,n1,replace=False)
    sample_mean=np.mean(sample)
    sample_var=np.var(sample)
    sample_min=np.min(sample)
    sample_max=np.max(sample)
    sample_range=sample_max-sample_min   
    samples_mean.append(sample_mean)
    samples_var.append(sample_var)
    samples_max.append(sample_max)
    samples_min.append(sample_min)
    samples_range.append(sample_range)
    i+=1
```
